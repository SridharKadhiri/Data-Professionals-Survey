```python
def replace_low_frequency(df, column_name, threshold=1):
    value_counts = df[column_name].value_counts()
    items_to_replace = value_counts[value_counts < threshold].index.tolist()
    df[column_name] = df[column_name].apply(lambda x: 'Other' if x in items_to_replace else x)
    return df
```

```python
# Cleaning technique to replace catogiries with low requency

def  replace_low_frequency_items(df,column , threshold = 1):
    item_count = df[column].value_counts()
    items_to_replace  = item_count[item_count<threshold].index.to_list()
    df[column]  = df[column].apply(lambda x:"Others" if x in items_to_replace else x)
    return df
```
